## Splitting the year across seasons
For a given year we split it into following seasons:  
Example: For 2022 bloom prediction,  
- __Spring__: April 2021, May 2021  
- __Summer__: June 2021, July 2021, August 2021  
- __Fall__: September 2021, October 2021, November 2021  
- __Winter__: Dec 2021, Jan 2022, Feb 2022, and first 7 days of March 2022  

## Features and what they capture
- __'r_chill_complete_doy'__: The day of the year when the chill requirement is met. Example: if for 2020 it is Jan 2 the value is 2 if it is Dec 31 (of the prev. year) the value is -1.  
- __'r_warm_completion'__: The amount of warm requirement completed until March 7 of the current year. Between 0.0 and 1.0.  
- __'year'__: The value of the year. Example: for 2022 it would be 2022.
- __'r_warm_29_completion'__: The amount of warm requirement completed on the 29th day of the current year. Between 0.0 and 1.0.    
- __'winter_chill'__: The number of days in the winter months (Dec, Jan, Feb, and first 7 days of March) below 5 degrees celcius.  
- __'winter_GDD'__: The number of growing degree-days in the winter months (Dec, Jan, Feb, and first 7 days of March) above 5 degrees celcius.  
- __'spring_GDD'__: The number of growing degree-days in the winter months (April, May) above 5 degrees celcius.   
- __'summer_avg'__: The average temperature in the summer months (June, July, August).
- __'fall_avg'__: The average temperature in the fall months (September, October, November).
- __'winter_avg'__: The average temperature in the winter months (Dec, Jan, Feb, and first 7 days of March).
- __'temperature_avg'__: The average temperature across all the months.
